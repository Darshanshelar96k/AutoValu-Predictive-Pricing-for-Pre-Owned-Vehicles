# outlier detection

```python
from scipy import stats
import numpy as np

# Calculate the z-scores
z_scores = stats.zscore(df)
z_scores
```

```python
# Convert to absolute values
abs_z_scores = np.abs(z_scores)

# Select data points with a z-scores above or below 3
filtered_entries = (abs_z_scores < 3).all(axis=1)

# Filter the dataset
df = df[filtered_entries]
df.shape
```

# creating multiple model

```python
from sklearn.linear_model import LinearRegression, Ridge,Lasso,ElasticNet
from sklearn.metrics import mean_squared_error
```

### LinearRegression

```python
lr = LinearRegression()
m_lr =lr.fit(x_train,y_train)
test_score = lr.score(x_test,y_test)
train_score = lr.score(x_train,y_train) 
print(train_score)
print(test_score)
```

```python
y_pred = m_lr.predict(x_test)
np.sqrt(mean_squared_error(y_pred,y_test))
```

### Ridge

```python
rg = Ridge()
m_rg =rg.fit(x_train,y_train)
test_score = rg.score(x_test,y_test)
train_score = rg.score(x_train,y_train) 
print(train_score)
print(test_score)
```

```python
y_pred = m_rg.predict(x_test)
np.sqrt(mean_squared_error(y_pred,y_test))
```

###  Lasso

### DecisionTreeRegressor

```python

from math import sqrt
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
m_dt = dt.fit(x_train,y_train)
test_score =dt.score(x_test,y_test)
train_score = dt.score(x_train,y_train)
print(train_score)
print(test_score)
y_pred = m_dt.predict(x_test)
score = dt.score(x,y)
print(score)
sqrt(mean_squared_error(y_pred,y_test))
```

###  Hyper parameter tuning for Decision Tree

```python
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV,KFold
# Define the Estimator and the Parameter Ranges
dt=DecisionTreeRegressor()
number_of_iterations = 5
min_samples_split = [2,5,7,10]
max_depth = range(2,30)
min_samples_leaf = [1,2,5]
# Define the param distribution dictionary
param_distributions = dict(min_samples_split=min_samples_split,
                           max_depth =max_depth,
                           min_samples_leaf =min_samples_leaf,
                          )

# Build the gridsearch
grid = RandomizedSearchCV(estimator=dt, 
                          param_distributions=param_distributions, 
                          n_iter=number_of_iterations, 
                          cv = 5)

grid_results = grid.fit(x_train, y_train)

# Summarize the results in a readable format
print("Best params: {0}, using {1}".format(grid_results.cv_results_['mean_test_score'], grid_results.best_params_))
results_df = pd.DataFrame(grid_results.cv_results_)
```

```python
result= grid_results.best_params_
min_samples_split = result.get("min_samples_split")
min_samples_leaf = result.get("min_samples_leaf")
max_depth = result.get('max_depth')
```

```python
dt = DecisionTreeRegressor(min_samples_split=min_samples_split,min_samples_leaf=min_samples_leaf,max_depth=max_depth)
dt_model = dt.fit(x_train,y_train)
test_score =dt.score(x_test,y_test)
train_score = dt.score(x_train,y_train)
print(train_score)
print(test_score)
score = dt.score(x,y)
print(score)
y_pred = dt_model.predict(x_test)
sqrt(mean_squared_error(y_pred,y_test))
```

### hyperparameter tuning for random forest

```python
# Define the Estimator and the Parameter Ranges
rf = RandomForestRegressor()
number_of_iterations = 10
min_samples_split = range(5,50)
max_depth = range(5,25)
max_features = range(5,13)
n_estimators = range(50,200)

# Define the param distribution dictionary
param_distributions = dict(min_samples_split=min_samples_split, 
                           max_depth=max_depth,
                           max_features=max_features,
                           n_estimators=n_estimators)

# Build the gridsearch
grid = RandomizedSearchCV(estimator=rf, 
                          param_distributions=param_distributions, 
                          n_iter=number_of_iterations, 
                          cv = 5)

grid_results = grid.fit(x_train, y_train)

# Summarize the results in a readable format
print("Best params: {0}, using {1}".format(grid_results.cv_results_['mean_test_score'], grid_results.best_params_))
results_df = pd.DataFrame(grid_results.cv_results_)
```

```python
result = grid_results.best_params_
n_estimators =result.get("n_estimators")
min_samples_split =result.get("min_samples_split")
max_features =result.get("max_features")
max_depth =result.get("max_depth")
```
